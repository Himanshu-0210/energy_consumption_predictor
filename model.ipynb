```python
import pandas as pd
import numpy as np
df=pd.read_csv('energydata_complete.csv')
```

```python
df.info()
```

```python
df['temperature']=df[['T1','T2','T3','T4','T5','T6','T7','T8','T9']].mean(axis=1)
df.drop(['T1','T2','T3','T4','T5','T6','T7','T8','T9'],axis=1,inplace=True)
```

```python
df['humidity']=df[['RH_1','RH_2','RH_3','RH_4','RH_5','RH_6','RH_7','RH_8','RH_9']].mean(axis=1)
df.drop(['RH_1','RH_2','RH_3','RH_4','RH_5','RH_6','RH_7','RH_8','RH_9'],axis=1,inplace=True)
```

```python
df.drop(['Press_mm_hg','Visibility','Tdewpoint','rv1','rv2','T_out','RH_out'],axis=1,inplace=True)
```

```python
df
```

```python
# Convert to datetime
df['date'] = pd.to_datetime(df['date'])

# Extract day
df['day'] = df['date'].dt.date

# Aggregate per day
df = df.groupby('day').agg({
    'Appliances': 'sum',      # total energy used in a day
    'temperature': 'mean',          # temperature
    'humidity': 'mean',         # humidity
    'Windspeed': 'mean',      
    'lights': 'sum'           # optional
}).reset_index()

# Rename columns for clarity
df.rename(columns={
    'Appliances': 'Energy_Consumption',
    'T_out': 'temperature',
    'RH_out': 'humidity'
}, inplace=True)
```

```python
df
```

```python
df['day'] = pd.to_datetime(df['day'])

df['day_of_week'] = df['day'].dt.dayofweek
df['is_weekend'] = df['day_of_week'].isin([5,6]).astype(int)
```

```python
df['Energy_Consumption'] = df['Energy_Consumption'] + df['lights']
```

```python
df.drop(['lights'],axis=1,inplace=True)
```

```python
df['Energy_Consumption'] = df['Energy_Consumption'] / 1000
```

```python
df['Appliance'] = (
    df['Energy_Consumption'] * 0.6 +
    df['temperature'] * 0.2 +
    df['humidity'] * 0.05
)

noise = np.random.randint(-2, 3, size=len(df))

df['Appliance'] = (df['Appliance'] + noise).round().astype(int)
df['Appliance'] = df['Appliance'].clip(lower=1)
```

```python
df
```

```python
df = df.sort_values(by='day')
```

```python
df['Energy_Consumption'] = df['Energy_Consumption'].clip(
    lower=df['Energy_Consumption'].quantile(0.01),
    upper=df['Energy_Consumption'].quantile(0.99)
)
```

```python
# Create strong lag features
df['lag_1'] = df['Energy_Consumption'].shift(1)
df['lag_2'] = df['Energy_Consumption'].shift(2)
df['lag_7'] = df['Energy_Consumption'].shift(7)

# Rolling mean
df['rolling_mean_3'] = df['Energy_Consumption'].rolling(3).mean()
df['rolling_mean_7'] = df['Energy_Consumption'].rolling(7).mean()

# Drop NA
df.dropna(inplace=True)
```

```python
df
```

```python
# Feature selection (optimized)
features = [
    'temperature',
    'humidity',
    'Windspeed',
    'Appliance',
    'day_of_week',
    'is_weekend',
    'lag_1',
    'lag_2',
    'lag_7',
    'rolling_mean_3',
    'rolling_mean_7'
]

X = df[features]
y = df['Energy_Consumption']
```

```python
# Time-series split
split = int(len(df) * 0.8)

X_train = X[:split]
X_test = X[split:]

y_train = y[:split]
y_test = y[split:]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

# Decision Tree
tree_reg = DecisionTreeRegressor(random_state=42)
tree_reg.fit(X_train, y_train)

# Random Forest
forest_reg = RandomForestRegressor(random_state=42)
forest_reg.fit(X_train, y_train)

#xgboost
xg=XGBRegressor(
    n_estimators=500,
    learning_rate=0.035,
    max_depth=5,
    subsample=0.9,
    colsample_bytree=0.9,
    gamma=0.03,
    reg_alpha=0.1,
    reg_lambda=1,
    random_state=42
)
xg.fit(X_train,y_train)


# Predict using training data
lin_preds = lin_reg.predict(X_test)
tree_preds = tree_reg.predict(X_test)
forest_preds = forest_reg.predict(X_test)
xg_preds=xg.predict(X_test)

# Calculate RMSE
lin_rmse = np.sqrt(mean_squared_error(y_test, lin_preds))
tree_rmse = np.sqrt(mean_squared_error(y_test, tree_preds))
forest_rmse = np.sqrt(mean_squared_error(y_test, forest_preds))
xg_rmse=np.sqrt(mean_squared_error(y_test, xg_preds))

lin_mae = mean_absolute_error(y_test, lin_preds)
tree_mae = mean_absolute_error(y_test, tree_preds)
forest_mae = mean_absolute_error(y_test, forest_preds)
xg_mae= mean_absolute_error(y_test, xg_preds)

lin_r2 = r2_score(y_test, lin_preds)
tree_r2 = r2_score(y_test, tree_preds)
forest_r2 =  r2_score(y_test, forest_preds)
xg_r2= r2_score(y_test, xg_preds)



print(f"Linear regression:\n mae:{lin_mae}\n rmse:{lin_rmse}\n r2:{lin_r2}")
print(f"Decision Tree:\n mae:{tree_mae}\n rmse:{tree_rmse}\n r2:{tree_r2}")
print(f"Random Forest:\n mae:{forest_mae}\n rmse:{forest_rmse}\n r2:{forest_r2}")
print(f"XGboost Regressor:\n mae:{xg_mae}\n rmse:{xg_rmse}\n r2:{xg_r2}")
```

```python
# Optimized XGBoost
from xgboost import XGBRegressor

model = XGBRegressor(
    n_estimators=500,
    learning_rate=0.035,
    max_depth=5,
    subsample=0.9,
    colsample_bytree=0.9,
    gamma=0.03,
    reg_alpha=0.1,
    reg_lambda=1,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
# Predictions
y_pred = model.predict(X_test)
```

```python
# Evaluation
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("MAE:", mae)
print("RMSE:", rmse)
print("R2 Score:", r2)
```
